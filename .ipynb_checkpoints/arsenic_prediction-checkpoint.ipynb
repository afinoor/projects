# Using Machine Learning to Aid Initiatives to Reduce Arsenic Poisoning
Between 20 and 45 million people in Bangladesh are exposed to arsenic contamination in drinking water, exceeding both national and WHO limits. The Bangladesh Arsenic Mitigation Water Supply Project (BAMWSP) reported arsenic presence in 61 out of 64 districts' tube wells in 2019-2020.

Van Geen's research in Araihazar, Bangladesh, demonstrated that simply testing well water and educating residents about arsenic's dangers could persuade 60% of them to switch to safer water sources. This method costs under 1 dollar per person, making it financially viable for developing countries compared to drilling private medium-depth wells, which cost 28 dollar per person.

Despite the low-cost and effectiveness, 40% of the population still continued using polluted wells. Most households in the Araihazar region are within 100 meters of a safe well, suggesting that awareness alone might not be enough for everyone.

This project aims to build a predictive model using demographic and behavioral data to identify those unlikely to switch to safer wells even after awareness campaigns. Four machine learning models will be evaluated to determine the best one for predicting this behavior, optimizing the approach to reduce arsenic poisoning in Bangladesh effectively.

Van Geen's Research: https://www.sciencedirect.com/science/article/abs/pii/S030438780600201X

Dataset used: http://www.stat.columbia.edu/~gelman/arm/examples/arsenic/wells.dat

Importing relevant libraries.

```python
import pandas as pd
#from google.colab import drive
import plotly.express as px
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
```

#Importing and Cleaning the Dataset

Importing the dataset from http://www.stat.columbia.edu/~gelman/arm/examples/arsenic/wells.dat

```python
df = pd.read_csv("http://www.stat.columbia.edu/~gelman/arm/examples/arsenic/wells.dat")

#splitting the single column in the dataset to 5 columns using space present
df = df['switch "arsenic" "dist" "assoc" "educ"'].str.split(" ",expand=True)

#removing the first column representing row number as pandas dataframe has indices by default
df = df.drop([df.columns[0]], axis = 1)

#renaming the columns
df.columns = ["switch", "arsenic", "dist", "assoc", "educ"]
df.head()
```

```python
#checking if there is any missing value
for column in df.columns.tolist():
  print(column)
  print(df[column].isnull().value_counts())
  print("")
```

```python
#changing all variables' data type to "float"
df = df.astype('float')
```

#Evaluating multiple machine learning models

```python
#creating a function to generate confusion matrix to evaluate model performance
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

##Data Preparation

Importing relevant libraries.

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from sklearn import metrics
import scipy.optimize as opt
from sklearn.metrics import f1_score, jaccard_score, log_loss, confusion_matrix, classification_report
import itertools
```

```python
X = df[['arsenic', 'dist', 'assoc', 'educ']].values
print(X[0:10])
print("")

y = df['switch'].values
print(y[0:5])
print("")

#normalize
X = preprocessing.StandardScaler().fit(X).transform(X.astype("float"))
X[0:5]
```

```python
#splitting the dataset into training and test datasets
#test to train dataset proportion would be 3:7
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 12)

print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

##K-Nearest Neighbours (KNN)

KNN, based on similarity of features, classifies the dataset into different groups (neighbors). The most crucial part of this model is choosing the optimum number of neighbors. Here from 1 to 9 neighbors will be chosen each time using a loop and the accuracy of the model will be measured. Therefore, optimum number of neighbors will be chosen based on the accuracy score.

```python
Ks = 15
mean_acc = np.zeros((Ks-1))

#creating a loop to build the KNN model each time with increasing number of neighbors
for n in range(1, Ks):

  KNN_model = KNeighborsClassifier(n_neighbors = n).fit(X_train, y_train)
  yhat = KNN_model.predict(X_test)

  #storing the accruacy score in an array
  mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)
```

```python
#creating a plot to visualize number of neighbors and their respective accuracy score
plt.plot(range(1,Ks), mean_acc)
print(mean_acc)
plt.gcf().set_size_inches(10, 6)
```

This graph shows, for number of neighbors starting from 0 to 9, accuracy for respective number of neighbors.

```python
#building a model with optimum number of neighbors
#mean_acc.argmax()+1 returns the optimum number of neighbors that has the highest accuracy
KNN_model = KNeighborsClassifier(n_neighbors = mean_acc.argmax()+1).fit(X_train, y_train)
yhat = KNN_model.predict(X_test)
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat, labels=[0, 1])
np.set_printoptions(precision=2)

print (classification_report(y_test, yhat))

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=["0", "1"], normalize= False,  title='Confusion matrix')
```

This confusion matrix has been constructed to visualize true positive, true negative, false positive, and false negative values predicted using the test dataset. From the matrix: 175 (True positive) + 362 (True Negative) datapoints were correctly predicted, and 205 (False positive) and 164 (False negative) datapoints were wrongly predicted. Using these values the f1 score has been calculated in the table.

Please note that if normalize parameter was selected to be true in the confusion matrix function, the matrix would have been built on the normalized data used in the machine learning models.

```python
KNN_jaccard_score = jaccard_score(y_test, yhat, pos_label = 1)
print("Jaccard score:", KNN_jaccard_score)
print("")

KNN_f1 = f1_score(y_test, yhat, average = "weighted")
print("f1 score:", KNN_f1)
print("")
```

Jaccard index shows that there is a 49.5% similarity between actual values of dependent variable in test dataset and our predicted values using the independent variables in test dataset.
F1 score, which takes into account false negative and false positive as well in the accuracy, is a better metric compared to mere accuracy and it ranges from 0 to 1 (1 being the score of a perfect model)

#Logistic Regression

By measuring probabilities, Logistic Regression model will predict whether a family witched to a new well or not.

```python
#importing relevant library and building logistic regression model
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(C = 0.01, solver = 'liblinear').fit(X_train, y_train)
yhat = LR.predict(X_test)
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat, labels=[0, 1])
np.set_printoptions(precision=2)

print (classification_report(y_test, yhat))

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=["0", "1"], normalize= False,  title='Confusion matrix')
```

```python
LR_jaccard_score = jaccard_score(y_test, yhat, pos_label=1)
print("Jaccard score:", LR_jaccard_score)
print("")

LR_f1 = f1_score(y_test, yhat, average = "weighted")
print("f1 score:",LR_f1)
print("")

yhat_prob = LR.predict_proba(X_test)
print(yhat_prob[0:5])
print("")
LR_log_loss = log_loss(y_test, yhat_prob)
print("log_ loss score:", LR_log_loss)
```

Los loss is a classification metric that measures performance on basis of the probabilities calculated by the model. In this model los loss metric has been used only for the Logistic Regression, as only this model in this project deals with probabilities.

# Decision Trees

```python
#importing libraries for decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
#building the predictive model based on decision tree
tree_clf = DecisionTreeClassifier(criterion= "entropy", max_depth = 4).fit(X_train, y_train)
yhat = tree_clf.predict(X_test)
```

```python
#importing relevant libraries to plot the tree
from  io import StringIO
import pydotplus
import matplotlib.image as mpimg
from sklearn import tree
```

```python
#plotting the decision tree made by the training dataset
dot_data = StringIO()
filename = "drugtree.png"
featureNames = df.drop("switch", axis = 1).columns.astype("string")
out=tree.export_graphviz(tree_clf,feature_names=featureNames,
                         out_file=dot_data,
                         class_names= np.unique(y_train).astype("str"),
                         filled=True,
                         special_characters=True,
                         rotate=False)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png(filename)
img = mpimg.imread(filename)
plt.figure(figsize=(100, 200))
plt.imshow(img,interpolation='nearest')
```

This tree shows the decision making criteria using the independent variables spread out in 4 layers.

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat, labels=[0, 1])
np.set_printoptions(precision=2)

print (classification_report(y_test, yhat))

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=["0", "1"], normalize= False,  title='Confusion matrix')
```

```python
tree_jaccard_score = jaccard_score(y_test, yhat, pos_label=1)
print("Jaccard score:", tree_jaccard_score)
print("")

tree_f1 = f1_score(y_test, yhat, average = "weighted")
print("f1 score:", tree_f1)
print("")
```
